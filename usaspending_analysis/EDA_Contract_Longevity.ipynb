# Exploratory Data Analysis: IT Contract Longevity
This notebook analyzes the downloaded longitudinal dataset (FY18-FY24) of federal IT contracts to assess duration, financial volume, and sub-contractor dependencies.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set aesthetic parameters for visualizations
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. Data Ingestion & Cleaning

```python
# Load the dataset
df = pd.read_csv('data/longitudinal_it_contracts_fy18_fy24.csv')

# Convert dates to datetime objects
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['End Date'] = pd.to_datetime(df['End Date'])

# Calculate base contract duration in days
df['Duration (Days)'] = (df['End Date'] - df['Start Date']).dt.days

# Clean Award Amount (ensure it's numeric)
df['Award Amount'] = pd.to_numeric(df['Award Amount'], errors='coerce')

# Fill NaNs in Subaward Count to differentiate Primes with and without Subcontractors
df['Subaward Count'] = df['Subaward Count'].fillna(0)
df['Has Subawards'] = df['Subaward Count'] > 0

# Preview the cleaned engineering frame
df[['Recipient Name', 'Award Amount', 'Duration (Days)', 'Subaward Count']].head()
```

## 2. Base Duration vs Award Amount Analysis

```python
# Plotting the distribution of contract durations
plt.figure(figsize=(12, 5))
sns.histplot(df['Duration (Days)'].dropna(), bins=50, kde=True)
plt.title('Distribution of Federal IT Contract Durations (Days)')
plt.xlabel('Duration in Days')
plt.ylabel('Frequency')
plt.show()
```

## 3. Sub-Contractor Dependency Comparison

```python
# Comparing contract length: Primes with Subcontractors vs Primes executing solo
plt.figure(figsize=(10, 6))
sns.boxplot(x='Has Subawards', y='Duration (Days)', data=df)
plt.title('Contract Longevity: Solo Primes vs. Primes with Subcontractors')
plt.xlabel('Utilized Subcontractors')
plt.ylabel('Duration (Days)')
plt.show()
```
